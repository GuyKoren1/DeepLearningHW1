```python
df_train = df[:463715]
df_test = df[463715:]

# convert to numpy
train_xs = df_train[x_labels].to_numpy()
train_ts = df_train[t_label].to_numpy()
test_xs = df_test[x_labels].to_numpy()
test_ts = df_test[t_label].to_numpy()
```

Q: Explain why it would be problematic to have some songs from an artist in the training set, and other songs from the same artist in the test set.



A: The model uses the training set to adjust it's NL varibles to give the best result for the those training samples.
If we wont maintain the training set and the test set completly disjointed, we will end up with the wrong model accrucy.
As our model in going over the test set, it will predict correctly the samples which apper in the training test as well, causing our models prediction to be higher than it realy is.

```python
feature_means = df_train.mean()[1:].to_numpy() # the [1:] removes the mean of the "year" field
feature_stds  = df_train.std()[1:].to_numpy()

train_norm_xs = (train_xs - feature_means) / feature_stds
test_norm_xs = (test_xs - feature_means) / feature_stds
```
